```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
Heroes_of_Pymoli = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(Heroes_of_Pymoli)

#Display the top 10 rows only

purchase_data_df.head(10)
```

## Player Count

* Display the total number of players


```python
Total_Players = len(purchase_data_df['SN'].unique())
Total_df = pd.DataFrame ({"Total Players":[Total_Players]})
Total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of Unique Items
unique_item_count=len(purchase_data_df['Item ID'].unique())

#Average Purchase Price
avg_purchase_price=purchase_data_df['Price'].mean()

#Total Number of Purchases
total_purchases=purchase_data_df['Purchase ID'].count()

#Total Revenue
total_revenue=purchase_data_df['Price'].sum()
```

```python
#Create a summary Data Frame to hold the results
purchase_summary_df=pd.DataFrame({"Number of Unique Items": [unique_item_count], "Average Price": [avg_purchase_price], 
                                  "Number of Purchases": [total_purchases], "Total Revenue": [total_revenue]})

#Give the displayed data cleaner formatting
purchase_summary_df['Average Price']=purchase_summary_df['Average Price'].map('${:.3}'.format)
purchase_summary_df['Total Revenue']=purchase_summary_df['Total Revenue'].map('${:,.6}'.format)
center_aligned_df = purchase_summary_df.style.set_properties(**{'text-align': 'center'})

#Display the summary data frame
center_aligned_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Percentage and Count of Male Players, Female and Other/Non-Disclosed
Gender_group = purchase_data_df.groupby("Gender")
Gender_unique= Gender_group['SN'].nunique()

Gender_percentage=(Gender_unique/Total_Players)*100

#Create Data Frame to Display Results
Gender_summary=pd.DataFrame({"Total Counts": Gender_unique, "Percentage of Players": Gender_percentage})
Gender_summary.index.name=None
Gender_summary["Percentage of Players"]=Gender_summary["Percentage of Players"].map("{:.2f}%".format)

Gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calculate Purchase Count
PA_Gender_group = purchase_data_df.groupby("Gender")
PA_Gender_count= PA_Gender_group['Purchase ID'].count()

#Average Purchase Price
PA_Gender_ppgroup = purchase_data_df.groupby("Gender")
PA_Gender_ppavg= PA_Gender_ppgroup['Price'].mean()

#Total Purchase Value
PA_Gender_pvgroup = purchase_data_df.groupby("Gender")
PA_Gender_pvavg= PA_Gender_pvgroup['Price'].sum()

#Average Total Purchase/Person
PA_Gender_appgroup = purchase_data_df.groupby("Gender")
PA_Gender_appavg= (PA_Gender_pvavg/Gender_unique)

#Create Data Frame to Display Results
PA_Gender_summary=pd.DataFrame({"Purchase Count": PA_Gender_count, "Average Purchase Price": PA_Gender_ppavg, "Total Purchase Value": PA_Gender_pvavg, "Avg Total Purchase per Person": PA_Gender_appavg})
PA_Gender_summary.index.name=None
PA_Gender_summary["Average Purchase Price"]=PA_Gender_summary["Average Purchase Price"].map("${:.2f}".format)
PA_Gender_summary["Total Purchase Value"]=PA_Gender_summary["Total Purchase Value"].map("${:,.2f}".format)
PA_Gender_summary["Avg Total Purchase per Person"]=PA_Gender_summary["Avg Total Purchase per Person"].map("${:.2f}".format)
center_aligned_df = PA_Gender_summary.style.set_properties(**{'text-align': 'center'})

#Display the Summary Data Frame
center_aligned_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish Bins for Ages

age_bins=[0, 9.90, 14.00, 19.90, 24.90, 29.90, 34.90, 39.90, 9999]
bin_names=["<10", "10-14", "15-19","20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins
purchase_data_df["Age Demographics Summary"]=pd.cut(purchase_data_df["Age"],age_bins, labels=bin_names)
purchase_data_df

#Calculate the numbers and percentages by age group
Age_Demographics_Group=purchase_data_df.groupby("Age Demographics Summary")
Age_unique= Age_Demographics_Group['SN'].nunique()

Age_percentage=(Age_unique/Total_Players)*100

#Create a summary data frame to hold the results
Age_Demo=pd.DataFrame({"Total Purchase Count": Age_unique,"Percentage of Players": Age_percentage})
Age_Demo.index.name=None

#Display Age Demographics Table
Age_Demo["Percentage of Players"]=Age_Demo["Percentage of Players"].map("{:.2f}%".format)
center_aligned_df = Age_Demo.style.set_properties(**{'text-align': 'center'})
center_aligned_df
```
